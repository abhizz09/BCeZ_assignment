### (a) Make a DataFrame of all the countries inside the 'Americas' continent. Print the size of the DataFrame and the name list of the countries.

```python
gapminder_sub_set_Americas.size
```

```python
country_in_america=gapminder_sub_set_Americas.country.unique()
```

### (b) Using a loop, calculate the average gdpPercap for all the countries and store them in an array.

```python
avgGDP=[""]*len(country_in_america)
for i in range (len(country_in_america)):
    x=gapminder[gapminder.country==country_in_america[i]]
    avgGDP[i]=x.gdpPercap.mean()
avgGDP    
```

```python
avgGDP=pd.DataFrame(avgGDP)
country_in_america=pd.DataFrame(country_in_america)
print(type(country_in_america))
#y1=[country_in_america,avgGDP]

y1=pd.concat((country_in_america, avgGDP),axis=1)
y1.columns = ['country', 'Average']
y1.set_index('country', inplace=True)
y1.head()
```

### (c) Draw a bar plot of the array in (b) where the x axis shows the names of the countries below the bars and y axis is the average gdpPercap. Axes should be labeled properly and the figure must have a title.
