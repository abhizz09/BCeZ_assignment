### Reading and Cleaning the Test Data 

```python
t_test=t_test.drop("Cabin",axis=1)
```

```python
t_test.dropna(inplace=True)
```

```python
t_train["Sex"].replace("male",0,inplace=True)
t_train["Sex"].replace("female",1,inplace=True)
t_train.Sex
```

```python
attri=t_train.drop(["Survived","Name","Ticket","Embarked"],axis=1)
classs=t_train.Survived
```

```python
t_test.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
attri_train,attri_test,classs_train,classs_test=train_test_split(attri,classs,test_size=0.35)
```

```python
print(attri_train)
```

### Applying machine learning algoriths

```python
#1> Decision Tree Classifier

from sklearn.tree import DecisionTreeClassifier
tr=DecisionTreeClassifier(max_depth=5)
tr.fit(attri_train,classs_train)
```

```python
from sklearn.metrics import accuracy_score
tr1=tr.predict(attri_test)
```

```python
accuracy_score(classs_test,tr1)*100
```
